```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('retail_sales_data_extended.csv')

# Add a Total_Sale column
df['Total_Sale'] = df['Quantity'] * df['Price_per_Unit']

# Show first few rows
print("ðŸ“Š First 5 Rows of Data:")
print(df.head())

# Group by Product
top_products = df.groupby('Product')['Total_Sale'].sum().sort_values(ascending=False)

# Group by Category
top_categories = df.groupby('Category')['Total_Sale'].sum().sort_values(ascending=False)

# Show Top 5 Products
print("\nðŸ”¥ Top 5 Products by Sales:")
print(top_products.head())

# Show Top 5 Categories
print("\nðŸ“ˆ Top 5 Categories by Sales:")
print(top_categories.head())

# Plot Top 5 Products
plt.figure(figsize=(10,5))
top_products.head(5).plot(kind='bar', color='skyblue')
plt.title("Top 5 Products by Total Sales")
plt.ylabel("Total Sales (â‚¹)")
plt.xlabel("Product")
plt.grid(axis='y')
plt.show()
```
